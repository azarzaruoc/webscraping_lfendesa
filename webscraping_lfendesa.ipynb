# 1. Importamos librerías

```python
import requests
from bs4 import BeautifulSoup
import csv
from datetime import datetime
```

```python
# Ruta de la página web
url = 'https://www.lfendesa.es/Inicio.aspx'

# Realizamos la petición a la web
req = requests.get(url)

# Obtenemos
statusCode = req.status_code
htmlText = req.text
```

```python
if statusCode == 200:
  # Pasamos el contenido HTML de la web a un objeto BeautifulSoup()
  html = BeautifulSoup(req.text, "html.parser")

  # Obtenemos el div donde se situan los enlaces del menu de navegación
  menu = html.find('div', {'class': 'content-nav'})

  # Obtenemos los links del menu 
  links = menu.find_all('a')

  # Guardamos los links en un diccionario
  enlacesMenu = {}
  for link in links:
    enlacesMenu[link['id']] = link['href']
  print(enlacesMenu)
else:
    print("Status Code %d" % statusCode)
```
